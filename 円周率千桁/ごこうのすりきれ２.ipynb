```python
import math
```

```python
def T(n, P, C):
    return math.pow(-1, n) * (P / C)
```

```python
def P(n, y):
    return 1/ math.pow(y, 2*n+1)
```

```python
def C(n):
    return 2*n + 1
```

```python
# 1. input num

y = 5.0
x = 1.0 / 5.0

# 2. init p,c,t,s

p = y/(y*y) 
c = 1.0
t = p/c
s = t
print("{n} is {s}".format(n=0, s=s))

# 23. update p,c,t,s
for n in range(1, 10):
    p /= y**2
    c += 2
    t = T(n, p, c)
    s += t
    print("{n} is {s}".format(n=n, s=t))
print("{n} is {s}".format(n=0, s=s))
```

```python
# my arctan2

def myarctan2(X, e=1.0):
    arctan = 0
    flag = 2
    N = 10
    for n in range(1, N+1, 2):
        buf = (10**e) // (((-1)**flag) * n * (X**n))
        arctan += buf
        flag += 1
        #print("{n} is {at}".format(n=(n+1)/2, at=buf))
    print("input:{x}, output:{o}".format(x=X, o=arctan))
    return arctan
```
